```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
duplicate_scala = pd.read_csv("/content/gdrive/MyDrive/GMCDE/experiments/sanity/2022-01-11-20-27_Duplicate_scala/Duplicate_scala.csv")
duplicate_python = pd.read_csv("/content/gdrive/MyDrive/GMCDE/experiments/sanity/2021-12-26-15-42_Duplicate_python/Duplicate_python.csv")
dilute_scala = pd.read_csv("/content/gdrive/MyDrive/GMCDE/experiments/sanity/2022-01-11-20-27_Dilute_scala/Dilute_scala.csv")
dilute_python = pd.read_csv("/content/gdrive/MyDrive/GMCDE/experiments/sanity/2021-12-26-13-56_Dilute_python/Dilute_python.csv")
```

```python
x_duplicate = duplicate_scala["duplicate"].to_numpy()
y_duplicate_GMCDE = duplicate_scala["gc"].to_numpy()
y_duplicate_dHSIC = duplicate_python["gc"].to_numpy()
```

```python
x_dilute = dilute_scala["dilution"].to_numpy()
y_dilute_GMCDE = dilute_scala["gc"].to_numpy()
y_dilute_dHSIC = dilute_python["gc"].to_numpy()
```

```python
f, axarr = plt.subplots(2,1, figsize = (8,8), sharex=True, sharey=False)
axes = axarr.flatten()
plt.sca(axes[0])
plt.plot(x_duplicate, y_duplicate_GMCDE, label = "GMCDE", color = "red")
plt.ylabel("Generalized Contrast (GC)")
plt.ylim(0.90, 1.10)
plt.title("Generalized Contrast (GC) with increasing duplicates in dimension", fontsize = 12)
plt.legend()

plt.sca(axes[1])
plt.plot(x_duplicate, y_duplicate_dHSIC, label = "dHSIC", color = "blue");
plt.xlabel("number of duplicates")
plt.ylabel("Generalized Contrast (GC)")
plt.ylim(0.0, 0.2)
plt.legend()
```

```python
f, axarr = plt.subplots(2,1, figsize = (8,8), sharex=True, sharey=False)
axes = axarr.flatten()
plt.sca(axes[0])
plt.plot(x_dilute, y_dilute_GMCDE, label = "GMCDE", color = "red");
plt.ylabel("Generalized Contrast (GC)")
plt.ylim(0.35, 1.05)
plt.title("Generalized Contrast (GC) with increasing dilution by independent dimensions", fontsize = 12)
plt.legend()
plt.sca(axes[1])
plt.plot(x_dilute, y_dilute_dHSIC, label = "dHSIC", color = "blue");
plt.xlabel("number of independent dimensions")
plt.ylabel("Generalized Contrast (GC)")
plt.ylim(0.0, 0.13)
plt.legend()
```
